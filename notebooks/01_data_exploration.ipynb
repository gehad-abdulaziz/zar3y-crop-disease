# 01. Data Exploration - Zar3y Project
This notebook explores the PlantVillage dataset subset for Egyptian crops.

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
from src.settings import DATA_DIR

# Check class distribution
classes = os.listdir(os.path.join(DATA_DIR, 'plant_village/train'))
counts = {c: len(os.listdir(os.path.join(DATA_DIR, 'plant_village/train', c))) for c in classes}
pd.Series(counts).plot(kind='bar')
plt.title('Class Distribution')
plt.show()
```
